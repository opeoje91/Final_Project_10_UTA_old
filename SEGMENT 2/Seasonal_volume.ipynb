```python
# File to load
AAPL_load = "AAPL_Pre_Official_DB_7_31.csv"

#Read data
AAPL_df = pd.read_csv(AAPL_load)
AAPL_df.head()
```

```python
AAPL_years = AAPL_df.Year.unique()
AAPL_years
```

```python
AAPL_summer_df = AAPL_df[AAPL_df["Season"] == "Summer"]
AAPL_summer_df.head()
```

```python
AAPL_fall_df = AAPL_df[AAPL_df["Season"] == "Fall"]
AAPL_fall_df.head()
```
